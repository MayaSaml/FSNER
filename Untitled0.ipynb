```python
from transformers import pipeline

generator = pipeline("text-generation", model="tiiuae/falcon-rw-1b")  # small 1B model

prompt = "Tell me something about the moon."
output = generator(prompt, max_new_tokens=50)

print(output[0]["generated_text"])
```

```python
from transformers import pipeline
```

```python
pipe = pipeline(
    "text-generation",
    model="HuggingFaceH4/zephyr-7b-beta",
    device_map="auto"
)

prompt = "<|system|>You are a helpful assistant.<|user|>Tell me something about the Moon.<|assistant|>"

response = pipe(prompt, max_new_tokens=100, do_sample=True, temperature=0.7)
print(response[0]["generated_text"])
```
